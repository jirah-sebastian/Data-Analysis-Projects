```python
import pandas as pd
import numpy as np
```

*I saw that there was question mark characters entered in the dataset so I replaced them with NaN, and I will process those null values later*

```python
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vR4ziT6AJ2_xaImeVtmASW8i0nH1No-qQ1tI5sm83aFtZywflKyLSmyS1fWo83Qm7-z8UQvR6r0gVab/pub?output=csv'
# reading the dataset
df = pd.read_csv(url,na_values="?")
print(df)
```

*In the above output we can see that the question marks have been replaced with NaN*

```python
df.info()
```

```python
df.describe()
```

```python
dtypes = df.dtypes

obj_cols = dtypes[dtypes == 'object'].index
int_cols = dtypes[dtypes == 'int'].index
print(int_cols, "\n")
print(obj_cols)
```

*Here it shows that there are 4262 empty values (NaN) in the dataframe*

```python
df.isna().sum().sum()
```

```python
# Here I will replace all NaN values with empty strings
df = df.apply(lambda x: pd.Series(x.dropna().values))
print(df)
```

```python
df = df.fillna('')

print(df) # Here we can see that NaN values are gone
```

*There are 24 duplicated values*

```python
# checking if there are duplicates
df.duplicated().sum()
```

*Here we remove them and also checked and it shows that there are no duplicated values anymore*

```python
# here I am going to remove duplicated values
DataFrame = df.drop_duplicates()
# Let's check if there are still duplicated values
DataFrame.duplicated().sum()
```
